```python
from datetime import datetime
# old=pd.read_csv("D:/DataSci/Projects/COVID/covid19_data.csv")
df=pd.read_csv("D:/DataSci/Projects/COVID/covid19_data2.csv")
#add in the date from one of the pages & extend 
# df.rename(columns={"Report_Date_x":"Report_Date"},inplace=True)
df["Report_Date"]=pd.to_datetime(df.Report_Date_x)
#.dt.strftime("m%/%d/%Y")
df.head(30)
```

```python
mid=df.filter(regex="UK")
if mid.shape[1]!=3:
    ch2=mid.filter(regex="cases")
    ch3=ch2.transpose().reset_index().rename(columns={"index":"Country"})
    #value to restrict on
    ch4=ch3.Country[ch3.iloc[:,1]==ch3.iloc[:,1].max()]
    dt=mid.filter(regex="deaths")
    dt2=dt.transpose().reset_index().rename(columns={"index":"Country"})
    #value to restrict on
    dt3=dt2.Country[dt2.iloc[:,1]==dt2.iloc[:,1].max()]
    
    rec=mid.filter(regex="deaths")
    rec2=rec.transpose().reset_index().rename(columns={"index":"Country"})
    #value to restrict on
    rec3=rec2.Country[rec2.iloc[:,1]==rec2.iloc[:,1].max()]
    mid=pd.concat([df[ch4],df[dt3],df[rec3]],axis=1)
mid
```

```python
import plotly.express as px
base=pd.DataFrame()
conts=["Germany","UK","France","USA", "Italy","Spain","Switzerland","Austria","Canada","Belgium"]
for col in conts:
    mid=df.filter(regex=col)
    if mid.shape[1]!=3:
        ch2=mid.filter(regex="cases")
        ch3=ch2.transpose().reset_index().rename(columns={"index":"Country"})
        #value to restrict on
        ch4=ch3.Country[ch3.iloc[:,1]==ch3.iloc[:,1].max()]
        dt=mid.filter(regex="deaths")
        dt2=dt.transpose().reset_index().rename(columns={"index":"Country"})
        #value to restrict on
        dt3=dt2.Country[dt2.iloc[:,1]==dt2.iloc[:,1].max()]

        rec=mid.filter(regex="deaths")
        rec2=rec.transpose().reset_index().rename(columns={"index":"Country"})
        #value to restrict on
        rec3=rec2.Country[rec2.iloc[:,1]==rec2.iloc[:,1].max()]
        mid=pd.concat([df[ch4],df[dt3],df[rec3]],axis=1)
    
    mid.columns=["Cases","Deaths", "Recovered"]
    mid2=pd.concat([df[["Report_Date"]],mid],axis=1)
    mid2["Country"]=col
    mid2.sort_values(by=["Country","Report_Date"],inplace=True)
    mid2["Incremental_Cases"]=mid2.Cases.diff()
    mid2["CaseRate"]=mid2.Incremental_Cases/(mid2.Cases-mid2.Incremental_Cases)
    mid2["Incremental_Deaths"]=mid2.Deaths.diff()
    mid2["DeathRate"]=mid2.Incremental_Deaths/(mid2.Deaths-mid2.Incremental_Deaths)
    mid2["RollingCases_last7"]=mid2.Incremental_Cases.rolling(7).mean()
    mid2["RollingDeaths_last7"]=mid2.Incremental_Deaths.rolling(7).mean()
    base=base.append(mid2)
fig=px.line(base,x="Report_Date",y="RollingCases_last7",color="Country")
fig.show()
```
